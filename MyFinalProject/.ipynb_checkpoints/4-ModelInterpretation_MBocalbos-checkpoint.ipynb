## Model Interpretation

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

X_train = pd.read_feather('./X_train.feather')
y_train = pd.read_feather('./y_train.feather')['shares']

X_val = pd.read_feather('./X_val.feather')
y_val = pd.read_feather('./y_val.feather')['shares']

rfr = RandomForestRegressor(n_estimators=42, max_features='log2', min_samples_leaf=46)
rfr.fit(X_train, y_train)
rfr.score(X_val, y_val)
```

```python
import eli5
from eli5.sklearn import PermutationImportance
pmi_limited = PermutationImportance(rfr).fit(X_val, y_val)
```

```python
eli5.explain_weights_df(pmi_limited, feature_names = X_val.columns.to_list())
```

We cannot conclude anything from the above graph about the number of shares because the number of videos in the online articles were mostly either 0 or 1. 

### PDP

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators=42, max_features='log2',min_samples_leaf=46)
rfr.fit(X_train, y_train)
```

```python
from pdpbox.pdp import pdp_plot, pdp_isolate
import pdpbox
def plot_pdp(model, X, feature_names, feature, clusters = None):
    iso = pdp_isolate(rfr, X_train, X_train.columns, feature)
    return pdp_plot(iso, feature, plot_lines = True, cluster=clusters, n_cluster_centers=clusters)
```

```python
plot_pdp(rfr, X_train.sample(2000), X_train.columns, 'num_videos', 15)
```

The plot above shows that the number of shares tend to increase as number of videos increases. As we can see, the shares increases at approximately 250 shares by having just one video on the article and then, it will increase little by little as it increase one unit at a time.

```python
plot_pdp(rfr, X_train.sample(2000), X_train.columns, 'num_hrefs', 15)
```

It is interesting to see that number of hrefs or hyperlinks have a negative impact on the number of shares at first. However, at a certain point, it will have a positive impact and will start increasing the number of shares at a slow rate.

```python
plot_pdp(rfr, X_train.sample(2000), X_train.columns, 'num_imgs', 15)
```

Another interesting insight is that the number of images have a negative impact on the number of shares.

### Working with a Single Tree

```python
first_tree = rfr.estimators_[0]
```

```python
from sklearn import tree
from IPython.display import SVG, display
from graphviz import Source 


def plot_tree(dt, feature_names, max_depth):
    return Source(tree.export_graphviz(dt, out_file=None,
                                feature_names=feature_names, max_depth=max_depth))
```

```python
plot_tree(first_tree, X_train.columns, 2)
```

```python
y_train.mean()
```

```python
first_obs = X_train.loc[0]
```

```python
import eli5
explained_df_first_obs = eli5.explain_prediction_df(first_tree, first_obs)

selected = explained_df_first_obs.feature.isin(['kw_avg_avg', '<BIAS>', 'num_videos',])
explained_df_first_obs[selected]
```

This shows the impact of kw_avg_avg and num_videos on our first tree. This gives us an idea that the kw_avg_avg increases the number of shares while in this first split, the number of videos negatively impacted our number of shares.

### Multiple Trees

```python
explained_df_first_obs_rfr = eli5.explain_prediction_df(rfr, first_obs)
explained_df_first_obs_rfr
```

The above table shows the average impact of our feature variables with our target variable. It tells us that the feature variables data_channel_missing, self_reference_min_shares, LDA_03, num_imgs, kw_min_avg, num_videos, n_unique_tokens, and self_reference_max_shares all have positive impact on the number of shares while num_hrefs, kw_avg_max, avg_negative_polarity, kw_avg_avg all have a negative impact. Two of these variables namely, num_hrefs and avg_negative_polarity to me really makes sense as having a negative effect on the number of shares.

```python
prediction_expls = []
for idx, row in X_val.iterrows():
    prediction_expls.append(eli5.explain_prediction_df(rfr, row))
```

```python
train_expl_df = pd.concat(prediction_expls, keys=X_train.index, 
                         names=['observation'])
```

```python
train_expl_df.iloc[:5]
```

```python
train_unstacked_df = (train_expl_df.groupby(['observation','feature'])
                               .weight
                               .aggregate('first')
                               .unstack())
train_unstacked_df.head()
```

```python
train_unstacked_df.describe()[train_unstacked_df.describe().index.isin(['mean', 'std'])]
```

```python
import seaborn as sns
fg = sns.lmplot(x='value', y='weight', col='feature',
                data=train_expl_df.loc[train_expl_df.feature.isin(['num_videos', 'num_imgs', 'num_hrefs'])], 
                col_order=['num_videos', 'num_imgs', 'num_hrefs'], sharex=False, sharey=False, col_wrap=3, fit_reg=False, 
                height=4, scatter_kws={'color':'salmon', 'alpha': 0.5, 's':30})


fg.fig.suptitle('Feature Contributions vs Feature Values (testing data)')
fg.fig.subplots_adjust(top=0.90);
```

From the chart above, it tells us that number of videos, number of images, and number of hrefs somehow has a positive linear correlation with number of shares. This is most especially obvious with the number of hrefs while the relationship is weaker with number of images.
